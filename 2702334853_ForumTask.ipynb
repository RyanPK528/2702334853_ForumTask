```python
# Name: Ryan Patrick Komala; ID: 2702334853
# Forum Task 1, The aim of this exercise is to make a gradebook for teacher's students.

# Dictionaries
lloyd = {
 "name": "Lloyd",
 "homework": [90.0,97.0,75.0,92.0],
 "quizzes": [88.0,40.0,94.0],
 "tests": [75.0,90.0]
}
alice = {
 "name": "Alice",
 "homework": [100.0, 92.0, 98.0, 100.0],
 "quizzes": [82.0, 83.0, 91.0],
 "tests": [89.0, 97.0]
}
tyler = {
 "name": "Tyler",
 "homework": [0.0, 87.0, 75.0, 22.0],
 "quizzes": [0.0, 75.0, 78.0],
 "tests": [100.0, 100.0]
}

# Functions
def average(numbers):
    total = float(sum(numbers))
    return total / len(numbers)

def get_average(student):
    homework = average(student["homework"])
    quizzes = average(student["quizzes"])
    tests = average(student["tests"])
    total_average = (homework * 10/100) + (quizzes * 30/100) + (tests * 60/100)
    return total_average

def get_letter_grade(score):
    if score >= 90 : 
        return "A"
    elif 80 <= score <= 90 : 
        return "B"
    elif 70 <= score <= 80 : 
        return "C"
    elif 60 <= score <= 70 : 
        return "D"
    else:
        return "F"

def get_class_average(students):
    results = []
    for student in students:
         results.append(get_average(student))
    return average(results)

# List
students = [lloyd, alice, tyler]

# Print 
print("""
        STUDENTS IN CLASS: 
""", end=" ")

for student in students:
    print(f"""
------------------------------------
Name: {student["name"]}
Homework: {student["homework"]}
Quizzes: {student["quizzes"]}
Tests: {student["tests"]}
------------------------------------
""")

class_average = get_class_average(students)
class_average_rounded = round(class_average, 2)
class_grade = get_letter_grade(class_average)

print(f"""
The average score (rounded to 2 decimal places) for the whole class is {class_average_rounded}.
Based on the class average, the class will be graded with a(n) {class_grade}.
""")

```

```python
# Name: Ryan Patrick Komala; ID: 2702334853
# Forum Task 2, The program should let the user enter a course number and, if the course is in the dictionary, it should display all of the course information. Otherwise, the program should display a message indicating that the course can’t be found.

# Tuple
course_numbers = ("COP 2510", "EGN 3000L", "MAC 2281", "MUH 3016", "PHY 2048")

# Dictionaries
course_name = {
    "COP 2510": "Programming Concepts",
    "EGN 3000L": "Foundations of Engineering Lab",
    "MAC 2281": "Calculus I",
    "MUH 3016": "Survey of Jazz",
    "PHY 2048": "General Physics I"
}

course_instructor = {
    "COP 2510": "Z. Beasley",
    "EGN 3000L": "J. Anderson",
    "MAC 2281": "A. Makaryus",
    "MUH 3016": "A. Wilkins",
    "PHY 2048": "G. Pradhan"
}

course_class_times = {
    "COP 2510": "MW 12:30pm – 1:45pm",
    "EGN 3000L": "TR 11:00am – 12:15pm",
    "MAC 2281": "MW 9:30am – 10:45am",
    "MUH 3016": "online asynchronous",
    "PHY 2048": "TR 5:00pm – 6:15pm"
}

# Function
def display_course_info (number):
    if number in course_numbers:
        print(f"""
The course details of {number} are:
1. Course Name: {course_name.get(number)}
2. Instructor: {course_instructor.get(number)}
3. Class Times: {course_class_times.get(number)} 
""") 
    
    else:
        print(f"{number} is an invalid course number.")


# Input
user_input = input("Enter a course number: ")
display_course_info(user_input)
```

```python
# Name: Ryan Patrick Komala; ID: 2702334853
# Forum Task 4, make a text to morse code converter using a tuple of the morse code

# Tuple for morse code
morse_code_tuple = (
    # A - Z
    '.-', '-...', '-.-.', '-..', '.', '..-.', '--.', '....', '..', '.---',
    '-.-', '.-..', '--', '-.', '---', '.--.', '--.-', '.-.', '...', '-',
    '..-', '...-', '.--', '-..-', '-.--', '--..', 
    # 0 - 9
    '.----', '..---', '...--', '....-', '.....',
    '-....', '--...', '---..', '----.', '-----',
    # Comma, Period, ?, Space
    '--..--', '.-.-.-', '..--..', ' ' 
)

# Function
def convert_to_morse(text):
    text = text.upper()
    morse_code = ""
    for char in text:
        if char.isalpha():
            morse_code += morse_code_tuple[ord(char) - ord('A')] 
        elif char.isnumeric():
            morse_code += morse_code_tuple[ord(char) - ord('0') + 25]
        elif char == ',':
            morse_code += morse_code_tuple[-4]
        elif char == '.':
            morse_code += morse_code_tuple[-3]
        elif char == '?':
            morse_code += morse_code_tuple[-2]
        elif char == ' ':
            morse_code += morse_code_tuple[-1]
        else:
            morse_code += ' '
    return morse_code

# Input and print
user_input = input("Enter a message: ")
morse_result = convert_to_morse(user_input)
print(f"The morse code for '{user_input}' is:")
print(f"{morse_result}")
```
